```python
#Import the needed libraries
import pandas as pd
import numpy as np
```

```python
#1) Get the data using the same process you did in Week 3.
df = pd.read_csv('labeledTrainData.tsv', sep='\t')
df.head(10)
```

```python
#Convert all text to lowercase letters.
df['review'] = df['review'].str.lower()
```

```python
#Remove line breaks <br>
def rem_br(txt):
    return txt.replace('<br>','').replace('<br />','')

df['review'] = df['review'].apply(rem_br)
```

```python
df
```

```python
#Remove punctuation and special characters from the text.
#df['review'] = df['review'].str.extract('(\w+)', expand = False)
df['review'] = df['review'].str.replace('[^\w\s]','')
```

```python
df
```

```python
#Remove stop words.
from nltk.corpus import stopwords
#nltk.download('stopwords')
from nltk.tokenize import word_tokenize


import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))


#Function to remove stop words from the string given
def rem_stops(txt):
    text_tokens = word_tokenize(txt)
    tokens_without_sw = [w for w in text_tokens if not w.lower() in stop_words]
    return str((" ").join(tokens_without_sw))


#Remove all the stop words from the reviews
df['rem_stop'] = df['review'].apply(rem_stops)
```

```python
df
```

```python
#Apply NLTK’s PorterStemmer.
from nltk.stem import *
from nltk.stem.porter import *
stemmer = PorterStemmer()

def get_singles(txt):
    splitted = txt.split(' ')
    singles = []
    for word in splitted:
        #return [stemmer.stem(plural) for plural in txt]
        singles.append(stemmer.stem(word))
    tmp_str = ''
    for word in singles:
        tmp_str += str(f'{word} ')
    return tmp_str[:-1]

df['stemmed'] = df['rem_stop'].apply(get_singles)
#df['review'] = np.vectorize(get_singles)(df['review'])
```

```python
#2) Split this into a training and test set.

train=df.sample(frac=0.8,random_state=200)
test=df.drop(train.index)
```

```python
#3) Fit and apply the tf-idf vectorization to the training set.
#Import the needed library
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

#X = review text, Y = positive or negative

tf_idf = TfidfVectorizer()

#applying tf idf to training data
X_train_tf = tf_idf.fit_transform(train['stemmed'])

#applying tf idf to training data
X_train_tf = tf_idf.transform(train['stemmed']).toarray()
```
