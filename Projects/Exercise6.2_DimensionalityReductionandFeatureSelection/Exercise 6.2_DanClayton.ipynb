```python
#Import the needed libraries
import pandas as pd
import numpy as np
```

```python
#1) Import the housing data as a data frame and ensure that the data is loaded properly.
df = pd.read_csv('train.csv')
df.head(10)
```

```python
#2) Drop the "Id" column and any features that are missing more than 40% of their values.

#Drop features with 40% missing values
limitPer = len(df) * .40001
culled_df = df.dropna(thresh=limitPer, axis=1)

#Drop ID
culled_df = culled_df.drop('Id', axis=1)
culled_df.head(10)
```

```python
#Build collection of numerical and quantitative columns for use elsewhere
quant_cols = ['MSSubClass','LotFrontage','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold','SalePrice']
qual_cols = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType', 'SaleCondition']

#Remove any columns that were removed earlier
for col in quant_cols:
    if col not in culled_df:
        quant_cols.remove(col)

for col in qual_cols:
    if col not in culled_df:
        qual_cols.remove(col)

print(quant_cols,'\n')
print(qual_cols,'\n')
print(culled_df.columns.tolist(),'\n')
```

```python
#3) For numerical columns, fill in any missing data with the median value.

for col in quant_cols:
    culled_df[col] = culled_df[col].fillna(culled_df[col].mean())
#culled_df=culled_df.fillna(df.mean())
culled_df.head(10)
```

```python
#4) For categorical columns, fill in any missing data with the most common value (mode).
qual_cols.remove('Fence')

for col in qual_cols:
    #Print the most common value
    most_pop = culled_df[col].value_counts().index[0]
    #print(culled_df[col].value_counts().index[0])
    culled_df[col] = culled_df[col].fillna(most_pop)
#df_clean = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
#5) Convert the categorical columns to dummy variables.

# Creating dummy variables:
df_dc = pd.get_dummies(culled_df, columns=qual_cols)
```

```python
#6) Split the data into a training and test set, where the SalePrice column is the target.
train=df_dc.sample(frac=0.8,random_state=200)
test=df_dc.drop(train.index)
#train
```

```python
test['SalePrice']
test.drop('SalePrice', axis=1)
```

```python
#7) Run a linear regression and report the R2-value and RMSE on the test set.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(train.drop(['SalePrice'],axis=1),train['SalePrice'])

r_sq = model.score(test.drop(['SalePrice'],axis=1),test['SalePrice'])
print('R2 value: ', r_sq)

#print(test)
#Predict outcomes
y_pred = model.predict(test.drop(['SalePrice'], axis=1))


print(y_pred)
print(test['SalePrice'])
rmse = mean_squared_error(test['SalePrice'], y_pred, squared=True)
print('RMSE: ', rmse)
```
