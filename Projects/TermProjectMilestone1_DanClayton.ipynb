What sets the price for anything that is bought or sold?  Some might say it's the seller who sets the price on the good or service in question, but that would only be half true.  Say, for example, that you are a fan of Martin Scorsese films, your favorite of which is "Goodfellas".  If you check Ebay right now (as of 6/28) you can find a mint condition "Goodfellas" poster, signed by Scorsese himself, as well as the 3 stars.  The price tag for such a treasure: $499 [plus $14.65 shipping]--what a bargain!  What is important to remember, though, is that the asking price on this gem is not really its price--as with any transaction a price hasn't really materialized until someone agrees to spend money on the item.  In another 3 days and 20 hours this auction will likely expire with no buyers proving an insurmountable difference between this item's asking price and its actual price.

Of course, the Ebay marketplace might not be the most efficient, so lets look at another busy marketplace: the stock market.  Stocks in major corporations tend to be highly liquid because there are many shares and they are actively traded by their owners.  At any given time you can lookup the bid and ask prices for a given stock to determine the stock's true price, which is usually somewhere between the bid/ask spread.  Because of its highly liquid nature traders and market makers tend to research the stocks that they own and the market tends to work efficiently to price in new information as it becomes available.  So why then, are people willing to pay $125 a share for a stock like GME (Game Stop), which hasn't had positive net income since 2018 when they could pay $20 a share for a company like AT&T, which earned $2.38 over the last 12 months?  The answer is that markets, while highly efficient, are not always rational.  

Because irrational behavior is, well, irrational, it can be hard to predict--like when a billionaire offers, out of the blue, to buy Twitter.  But what if there was a way to predict irrational market behavior?  What if you could tell beforehand that a stock was about to be skyrocketed by irrational investors?  You could then take the opportunity to buy that stock before its price gets driven up, selling it once the moment slows down and earning a tidy profit.  Gamestop (GME) is one of several, so called, "Meme Stocks", which irrational investors have chosen to buy and hold for reasons other than profit.  

The most popular platform for crowning meme stocks has been the Reddit page r/Wallstreetbets.  I intend to use the Reddit API to scrape ticker information and comments which will then be used to produce sentiment scores from r/Wallstreetbets posts to see if I can predict which are the next meme stocks.  If done correctly, then there is an opportunity to make money by buying meme stocks while they are still fairly valued.  If done incorrectly, then there is a good opportunity to loose money.

From looking at the charts below, one can see that a meme stock tends to skyrocket and then quickly adjust downward as investors come to their senses.  This can happen a number of times, but sustained price elevation is unlikely, because there is no intrinsic reason for the stock price to be elevated--thus any position would need to be quickly entered into and then closed out of.

```python
#Include some libraries that you might need
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
#Import the data set showing meme stock prices over the last 2 years
df = pd.read_csv('Meme_Prices_01012020-06282022.csv')
#Convert dates to date/time format
df['Date'] = pd.to_datetime(df['Date'])

df.head(10)
```

```python
#Graph Bed Bath and Beyond
plt = df[['Date', 'BBBY']].plot.line(y='BBBY', x='Date', rot=0, title='Bed Bath and Beyond PX Time\n(Consensus Price in Red)', legend=False)
plt.axhline(y=7.19, color='red', linestyle='--')

for i, t in enumerate(plt.get_xticklabels()):
    if (i % 2) != 0:
        t.set_visible(False)
```
